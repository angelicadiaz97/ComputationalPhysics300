```python
#let's plot the derivative of a function y=x**2 with domain [-5,5] letting h= 0.5 using all of the finite difference methods

import numpy as np 
import matplotlib.pyplot as plt

#define the function and parameters
f = lambda x: -x**4 + 4*x**3 + 2
df = lambda x: -4*x**3 + 12*x**2 
h = 0.5

#draw the plot
x = np.linspace(0,3,100) #100 points
plt.plot(x,df(x),"k")
# Forward difference  
dff1 = ((f(x+h) - f(x))/h)
plt.plot(x,dff1,'--r')
# Backward difference 
dff2 = (f(x) - f(x-h))/h
plt.plot(x,dff2, '--b')
# Central difference
dff3 = (f(x+h) - f(x-h))/(2*h)
plt.plot(x,dff3,'--g')

plt.xlabel('x')
plt.ylabel('f(x)') 
plt.legend(["theoretical f '(x)","Forward difference","Backward difference","Central difference"])
plt.grid()
plt.show()

#Notice that the central difference is closest to the theoretical as h-->0, the numerical solutions get closer to the analytical solutions 
```
